```python
import numpy as np
 
class DLNN:
    # Initialization function
    def __init__(self,input_array,output_array,layer_node_count_list,w_step_size=0.05,b_step_size = 0.05):
        self.layers = []    
        self.input_array = input_array
        print(self.input_array)
        self.output_array = output_array     
        print(self.output_array)
        self.w_step_size = w_step_size      
        self.b_step_size = b_step_size      
        self.layer_node_count_list = layer_node_count_list
        self.layers_count = len(layer_node_count_list)-1
        self.w_dic = {} 
        self.b_dic = {} 
        for i in range(1,len(layer_node_count_list)):
            self.w_dic[i-1] = 2*(np.random.random((layer_node_count_list[i-1],layer_node_count_list[i]))-0.5)   
            self.b_dic[i-1] = 2*(np.random.random((1,layer_node_count_list[i]))-0.5)    
        # print (self.w_dic)
        # print(self.b_dic)

        self.norm_max_array = None
        self.norm_min_array = None
 
    # Activation function
    # f(x) = 1/(1+e**(-x))
    def sigmoid(self,x,derivative=False):   
        if (derivative == True):
            return x * (1 - x)
        return 1 / (1 + np.exp(-x))
    def forward(self):
        self.layers = []    
        self.layers.append(self.input_array)  
        for i in range(self.layers_count):
            laywer_value_raw = np.dot(self.layers[i],self.w_dic[i]) + self.b_dic[i]
            laywer_value = self.sigmoid(laywer_value_raw)
            self.layers.append(laywer_value)
        # print(self.layers)
 


    def backword(self):
        delta_list = []
        theta_output = (self.output_array - self.layers[-1])*self.sigmoid(self.layers[-1],derivative=True) 
        delta_list.append(theta_output)
        for i in range(self.layers_count-1,0,-1):
            theta_middle = np.dot(delta_list[-1],self.w_dic[i].T) * self.sigmoid(self.layers[i],derivative=True)
            delta_list.append(theta_middle)

        delta_list.reverse()
        w_change_dic_1 = {}
        b_change_dic_1 = {}
        w_change_dic = {}
        b_change_dic = {}
        for i in range(len(delta_list)):
            w_change_dic_1[i] = np.dot(self.layers[i].T,delta_list[i]) * self.w_step_size
            b_change_dic_1[i] = np.dot(np.array([[1]]*len(self.input_array)).T,delta_list[i])*self.b_step_size
            w_change_dic[i] = np.dot(self.layers[i].T,delta_list[i]) * self.w_step_size
            b_change_dic[i] = np.dot(np.array([[1]]*len(self.input_array)).T,delta_list[i])*self.b_step_size
        for i in w_change_dic.keys():
            self.w_dic[i] += w_change_dic[i]
            self.b_dic[i] += b_change_dic[i]
 

    def check_data(self):
        input_shape = self.input_array.shape
        output_shape = self.output_array.shape
        # print(input_shape,output_shape)
        if len(input_shape) != 2:
            print("The input sample must be a 2-dimensional np sequence，Current sequence dimension：{0}！".format(len(input_shape)))
            return
 
        if len(input_shape) != len(output_shape): 
            if len(output_shape) == 1:
                tmp_output_array = np.array([self.output_array]).T
                print(":&gt;Output sequence by{0}change into:{1}".format(self.output_array,tmp_output_array))
                self.output_array = tmp_output_array
                output_shape = self.output_array.shape
 
        if input_shape[0] == output_shape[0]:
            pass
        else:
            if output_shape[0] == 1 and input_shape[0] == output_shape[1]:    #
                print(":&gt;Output sequence by{0}change into:{1}".format(self.output_array,self.output_array.T))
                self.output_array = self.output_array.T
                output_shape = self.output_array.shape
            else:
                print("The length of the input and output samples is different and cannot be processed further!")
                return
        if input_shape[1] != self.layer_node_count_list[0]:
            print("The number of input layer nodes is wrong, the input is：{0}，and based on the data (number of input features) should be{1}!".format(self.layer_node_count_list[0],input_shape[1]))
            return
        if output_shape[1] != self.layer_node_count_list[-1]:
            print("The number of input layer nodes is wrong, the input is：{0}，and based on the data (number of input features) should be{1}!".format(self.layer_node_count_list[-1],output_shape[1]))
            return
        print("-v- The data is correct and you can start the calculation.……")
 
 
    def preview(self,pre_input_array):
        input_data_shape = pre_input_array.shape   
        if len(input_data_shape) != 2:
            if len(input_data_shape) == 1:
                print("Forecast: Input data dimension is 1, try to correct...")
                pre_input_array = np.array([pre_input_array])
            else:
                print("Forecast: The input data shape is{0},Unable to fix, exit!".format(input_data_shape))
                exit(0)
        pre_layers = []    
        pre_layers.append(pre_input_array)  
        for i in range(self.layers_count):
            laywer_value_raw = np.dot(pre_layers[i],self.w_dic[i]) + self.b_dic[i]
            laywer_value = self.sigmoid(laywer_value_raw)
            pre_layers.append(laywer_value)
        print("Predicted output value:{0}".format(pre_layers[-1].T))
        return pre_layers
```
